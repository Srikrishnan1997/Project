```python
# Data analysis on the INDIA trade dataset 

import pandas as pd
from matplotlib  import pyplot as plt
# read the csv files of both import data and export data
df=pd.read_csv("C:/Users/srikrishnan.s/Desktop/DATASETS/2018-2010_export.csv")
df1=pd.read_csv("C:/Users/srikrishnan.s/Desktop/DATASETS/2018-2010_import.csv")
df
```

```python
# Group the dataset by HSC code and Year for both import and export data.
D=df.groupby(['HSCode','year']).sum()
D1=df1.groupby(['HSCode','year']).sum()
D1
```

```python
#Find the total import and export values by year
total_import_price=D1.groupby(['year']).sum()
total_export_price=D.groupby(['year']).sum()
#Find the difference of the export and import values
#If the differecnce value is  positive then the country has gained a profit throuugh trade
#If the differecnce value is  negative then the country has aquired a loss throuugh trade
df_add=total_export_price-total_import_price
df_add
```

```python
# Now let us find maximum export value of india between the year 2010 - 2018

max_value_export=df1['value'].max()
df1[df1['value']==max_value_export]
```

```python
# Now let us find  the maximum export value for each year from 2010-2018 and the corresponding commodity
DFF=df1.groupby(['year'])['value'].max()
x=[]
for i in DFF:
    x.append(i)
Max_export_values=df1.loc[df1['value'].isin(x)]
Max_export_values
```
